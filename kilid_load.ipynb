```python
convert_fac = {'آسانسور': "elevator", 
'آنتن مرکزی': "central antenna",
'استخر': "pool",
'انباری': "warehouse",
'بالکن': "balcony",
'تهویه مطبوع': "air condition",
'جکوزی': "jacuzzi",
'درب ریموت': "remote door",
'روف گاردن': "roof garden",
'سالن اجتماعات': "conference hall",
'سالن ورزش': "sports hall",
'سونا': "sauna",
'قدرالسهم': "qadrosahm",
'قیمت توافقی': "tavafoghi price",
'لابی': "lobby",
'مال': "mall",
'پاساژ': "mall",
'مشارکتی': "collaborative",
'معاوضه': "changable",
'موقعیت اداری': "administrative position",
'نگهبان': "guard",
'وام دار': "have l,oan",
'پیش فروش': "preorder",
}

kilid['facilities'] = kilid['facilities'].apply(lambda x: [convert_fac.get(item, item) for item in x])


```

```python
np.set_printoptions(threshold=np.inf)

def clean_and_convert(value):
    persian_numbers = '۱۲۳۴۵۶۷۸۹۰'
    english_numbers = '1234567890'
    for persian,english in zip(persian_numbers,english_numbers):
        value = value.replace(persian,english)
    return value
cols_to_numeric = ['meterage', 'age', 'parking', 'price']

kilid[cols_to_numeric] = kilid[cols_to_numeric].apply(clean_and_convert)
```

```python
kilid["price"] = kilid['price'].apply(lambda x : x.replace('هزار','Thousand'))
```

```python
kilid["price"] = kilid['price'].apply(lambda x : x.replace('میلیون','milion'))
```

```python
kilid["price"] = kilid['price'].apply(lambda x : x.replace('میلیارد','bilion'))
```

```python
kilid["properties"] = kilid['properties'].apply(lambda x : x.replace('مسکونی','residential'))
```

```python
kilid["type"] = kilid['type'].apply(lambda x : x.replace('مسکونی','residential'))
```

```python
kilid["func"] = kilid['func'].apply(lambda x : x.replace('آپارتمان/برج','apartment'))
```

```python
kilid.drop(['floor','date','type','title','desc'],inplace=True,axis=1)
```

```python
kilid['area'] = kilid['address'].apply(lambda x: re.findall('منطقه\s(\d*)',x)[0])
```

```python
kilid['address'] = kilid['address'].apply(lambda x: re.sub('منطقه\s\d*','',x))
```

```python
kilid['address'] = kilid['address'].apply(lambda x :x.replace('تهران ،',''))
```
